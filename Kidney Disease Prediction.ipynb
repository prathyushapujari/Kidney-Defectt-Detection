```python
import streamlit as st
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import KNNImputer
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.pipeline import Pipeline
import pickle

def header1(url): 
    st.markdown(f'<p style="color:#FF0000;font-size:50px;border-radius:2%;"><center><strong>{url}</strong></center></p>', unsafe_allow_html=True)

def header2(url): 
    st.markdown(f'<p style="color:#3CB043;font-size:50px;border-radius:2%;"><center><strong>{url}</strong></center></p>', unsafe_allow_html=True)

def predict():

    value_dict = { "NA":np.nan, "-": np.nan, "Present": 1, "Not Present": 0, "Good": 1,
                   "Poor": 0, "Normal": 1, "Abnormal": 0, "Yes": 1, "No": 0}

    test_input = []

    #st.title("Kidney Disease Prediction")

    st.warning("**NOTE** \n\n 1) If there is no data for a specific parameter type 'NA' \n\n 2) For drop down select   ' - '")

    col1, col2 = st.columns([1,1])

    with col1:
        age = st.text_input('Age',48)
        test_input.append(age)
        
    with col2:
        bp = st.text_input("Blood Pressure (bp)",80)
        test_input.append(bp)

    col1, col2 = st.columns([1,1])

    with col1:
        sg = st.text_input('Specific Gravity',1.020, key = "sg")
        test_input.append(sg)
        
    with col2:
        al = st.text_input("Albumin",1.0, key = "al")
        test_input.append(al)

    col1, col2 = st.columns([1,1])

    with col1:
        su = st.text_input('Sugar',0)
        test_input.append(su)
        
    with col2:
        bgr = st.text_input('Blood Glucose Random',121)
        test_input.append(bgr)

    col1, col2 = st.columns([1,1])

    with col1:
        bu = st.text_input('Blood Urea',36)
        test_input.append(bu)
        
    with col2:
        sc = st.text_input('Serum Creatinine',1.2)
        test_input.append(sc)

    col1, col2 = st.columns([1,1])

    with col1:
        sod = st.text_input('Sodium',"NA")
        test_input.append(sod)
        
    with col2:
        pot = st.text_input('Potassium',"NA")
        test_input.append(pot)

    col1, col2 = st.columns([1,1])

    with col1:
        hemo = st.text_input('Hemoglobin', 15.4)
        test_input.append(hemo)
        
    with col2:
        pcv = st.text_input('Packed Cell Volume',44)
        test_input.append(pcv)


    col1, col2 = st.columns([1,1])

    with col1:
        wbcc = st.text_input('White Blood Cell Count', 7800)
        test_input.append(wbcc)
        
    with col2:
        rbcc = st.text_input('Red Blood Cell Count', 5.2)
        test_input.append(rbcc)

    
    col1, col2 = st.columns([1,1])

    with col1:
        rbc = st.selectbox("Red Blood Cells", ['-','Normal','Abnormal'])
        test_input.append(rbc)
        
    with col2:
        pc = st.selectbox("Pus Cell", ['-','Normal','Abnormal'])
        test_input.append(pc)

    col1, col2 = st.columns([1,1])

    with col1:
        pcc = st.selectbox("Pus Cell Clumps", ['-','Present','Not Present'])
        test_input.append(pcc)
        
    with col2:
        ba = st.selectbox("Bacteria", ['-','Present','Not Present'])
        test_input.append(ba)

    col1, col2 = st.columns([1,1])

    with col1:
        htn = st.selectbox("Hypertension", ['-','Yes','No'])
        test_input.append(htn)
        
    with col2:
        dm = st.selectbox("Diabetes Mellitus", ['-','Yes','No'])
        test_input.append(dm)

    col1, col2 = st.columns([1,1])

    with col1:
        cad = st.selectbox("Coronary Artery Disease", ['-','Yes','No'])
        test_input.append(cad)
        
    with col2:
        appet = st.selectbox("Appetite", ['-','Good','Poor'])
        test_input.append(appet)

    col1, col2 = st.columns([1,1])

    with col1:
        pe = st.selectbox("Pedal Edema", ['-','Yes','No'])
        test_input.append(pe)
        
    with col2:
        ane = st.selectbox("Anemia", ['-','Yes','No'])
        test_input.append(ane)

    for i in range(0,14):
        if test_input[i] == "NA":
            test_input[i] = value_dict[test_input[i]]
        else:
            test_input[i] = float(test_input[i])

    for i in range(14,len(test_input)):
        test_input[i] = value_dict[test_input[i]]
    
    if st.button("Predict"):
        with open('ckd_prediction_model.pkl', 'rb') as f:
            model = pickle.load(f)

        output = int(model.predict([test_input])[0])

        col1, col2, col3 = st.columns([0.2,6,0.2])
        
        with col2:
            if output == 0:
                header1("Chronic Kidney Disease")

            else:
                header2("Not Chronic Kidney Disease")
        

if __name__ == "__main__":


    col1, col2 = st.columns([1,3])

    with col1:
        st.image("woxsen_logo.png")
    with col2:
        st.title("Kidney Disease Prediction")

    
        
    predict()

```
